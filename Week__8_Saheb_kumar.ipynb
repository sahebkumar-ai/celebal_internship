## Problem Statement
Build a **Single-Agent Smart Assistant** that:

### The agent should handle:

##Part 1: Single Agent Pipeline

````python
import logging
from datetime import datetime
````

# Logging Configuration

````python
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)
````

````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

# 🛠️ BONUS TOOL 3: Word Counter

````python
def word_count(text: str) -> int:
    """Count number of words."""
    try:
        return len(text.split())
    except Exception:
        return 0
````

# 🛠️ BONUS TOOL 4: Reverse Text

````python
def reverse_text(text: str) -> str:
    """Reverse the given text."""
    try:
        return text[::-1]
    except Exception:
        return ""
````

# 🛠️ BONUS TOOL 5: Current Date

````python
def current_date() -> str:
    """Return current date."""
    return datetime.now().strftime("%d-%m-%Y")
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)
#     # TODO: Implement routing logic

#     # Example structure:
#     # if "calculate" in query_lower:
#     #     ...
#     # elif "keywords" in query_lower:
#     #     ...
#     # else:
#     #     ...

def agent(query: str):

    query_lower = query.lower()

    logging.info(f"Received Query: {query}")

    try:

        # Calculator Routing

        if any(word in query_lower for word in ["calculate", "compute", "solve"]):

            expression = query_lower

            for word in ["calculate", "compute", "solve"]:
                expression = expression.replace(word, "")

            expression = expression.strip()

            result = calculator(expression)

            if result == "Error in calculation":

                logging.error("Calculation Error")

                return {
                    "type": "error",
                    "result": result
                }

            logging.info("Calculator Tool Executed")

            return {
                "type": "calculation",
                "result": result
            }


        # Keyword Extraction Routing

        elif any(word in query_lower for word in ["keyword", "keywords", "extract"]):

            text = query

            if "from" in query_lower:
                index = query_lower.find("from")
                text = query[index + len("from"):].strip()

            keywords = extract_keywords(text)

            logging.info("Keyword Tool Executed")

            return {
                "type": "keywords",
                "result": keywords
            }

        # Word Count Routing (Bonus)

        elif "count words" in query_lower:

            text = query_lower.replace("count words", "").strip()

            result = word_count(text)

            logging.info("Word Count Tool Executed")

            return {
                "type": "word_count",
                "result": result
            }

        # Reverse Text Routing (Bonus)

        elif "reverse" in query_lower:

            # text = query.replace("reverse", "").strip()

            # result = reverse_text(text)
            text = query.replace("Reverse", "").replace("reverse", "").strip()
            result = reverse_text(text)

            logging.info("Reverse Text Tool Executed")

            return {
                "type": "reverse",
                "result": result
            }

        # Date Routing (Bonus)

        elif "date" in query_lower:

            logging.info("Date Tool Executed")

            return {
                "type": "date",
                "result": current_date()
            }

        # General Response

        else:

            logging.info("General Response")

            return {
                "type": "general",
                "result": f"You asked: {query}"
            }

    except Exception as e:

        logging.error(str(e))

        return {
            "type": "error",
            "result": str(e)
        }
````

##1. Explain the concept of a stateful directed graph in agent pipelines. How does it differ from a simple linear pipeline?

A stateful directed graph is a workflow where each node can access and update shared state while processing a task. The graph consists of nodes connected by directed edges, allowing the workflow to branch, loop, or revisit previous steps based on conditions. This enables dynamic decision-making and memory across the execution. In contrast, a simple linear pipeline executes tasks in a fixed sequence from start to finish without branching, looping, or maintaining shared state, making it less flexible for complex workflows.

##2. Describe the role of nodes and edges in an agent workflow. Give an example of each.

Nodes represent the individual operations performed by an agent, such as analyzing a query, calling a tool, or generating a response. Edges define the flow of execution by connecting one node to another and determining the next step. For example, a Calculator Tool is a node, while the connection from the Intent Detection node to the Calculator Tool node is an edge that routes mathematical queries for calculation.

##3. What is conditional routing in an agent system? Design a simple rule-based routing logic for three different query types.

Conditional routing is the process of directing a user's request to different tools or processing paths based on its intent. A simple rule-based routing logic can be:

If the query contains "calculate", send it to the Calculator Tool.
If the query contains "keywords", send it to the Keyword Extraction Tool.
Otherwise, generate a General Response.

This ensures that each query is handled by the most appropriate component.

##4. Why are cycles (loops) important in agent pipelines? Provide a use case where a retry loop is necessary.

Cycles allow an agent to repeat a task until a desired condition is met. They improve robustness by enabling retries after failures or repeated refinement of outputs. For example, if an API request fails due to a temporary network issue, the agent can automatically retry the request several times before returning an error. This increases reliability without requiring user intervention.

##5. Explain how a single-agent system can simulate multi-agent behavior internally.

A single-agent system can simulate multi-agent behavior by dividing its workflow into specialized internal roles. For example, it may first analyze the user's intent, then select the appropriate tool, execute the task, and finally generate a response. Although only one agent performs all actions, each stage behaves like a specialized sub-agent, allowing the system to mimic a multi-agent architecture while remaining simpler to implement and maintain.

##6. What are JSON schema tools? How do they help in structuring tool inputs and outputs?

JSON schema tools define the expected structure, data types, and required fields for tool inputs and outputs. They validate data before processing, ensure consistent communication between the agent and external tools, reduce errors caused by invalid inputs, and make outputs predictable and easy to parse. This improves interoperability and reliability in agent systems.

##7. Compare sequential tool calls and parallel tool calls. When would you prefer one over the other?

Sequential tool calls execute one after another, where each step depends on the output of the previous step. They are suitable when tasks have dependencies, such as extracting data before summarizing it. Parallel tool calls execute multiple independent tasks simultaneously. They are preferred when tasks do not depend on one another, such as querying multiple APIs at the same time, reducing total execution time and improving efficiency.

##8. How would you implement error handling in a tool-using agent? Provide at least two strategies.

Error handling improves the reliability of an agent by managing failures gracefully. One strategy is using try-except blocks to catch exceptions and return meaningful error messages instead of crashing. Another strategy is implementing retry mechanisms that automatically repeat failed operations caused by temporary issues such as network failures. Additionally, logging errors helps developers monitor failures and simplify debugging.

##9. What is trajectory evaluation in agent systems? Why is it important beyond just checking final output?

Trajectory evaluation examines the complete sequence of actions an agent takes while solving a task, including reasoning steps, tool selection, intermediate outputs, and decisions. It is important because an agent may produce the correct final answer using inefficient or incorrect reasoning. Evaluating the full trajectory helps identify mistakes, optimize workflows, improve tool usage, and increase the overall reliability of the agent.

##10. Define task completion rate and cost metrics. How would you measure and optimize them in a real-world system?

Task completion rate measures the percentage of user requests successfully completed by the agent. Cost metrics represent the resources consumed during execution, such as API calls, computation time, token usage, or monetary expense. These metrics can be measured by monitoring agent performance across many tasks. They can be optimized by improving routing accuracy, minimizing unnecessary tool calls, caching repeated results, selecting efficient models, and reducing execution time while maintaining high accuracy.
